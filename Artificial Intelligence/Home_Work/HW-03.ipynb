# Завдання 1: Створення Series
- Створи Series з 5 чисел: [10, 20, 30, 40, 50]
- Присвой до кожного елемента індекси у вигляді днів тижня: ['Пн', 'Вт', 'Ср', 'Чт', 'Пт']
- Виведи значення по індексу 'Ср' Виведи усі індекси та значення окремо

```python
import pandas as pd
import numpy as np

data = [10, 20, 30, 40, 50]
week_days = ['Пн', 'Вт', 'Ср', 'Чт', 'Пт']
my_series = pd.Series(data, index=week_days)

print("Створення Series:")
print(my_series)
print("=="*20)

print(f"Значення за індексом 'Ср': {my_series['Ср']}")
print("=="*20)

print("Всі індекси Series:")
print(my_series.index)
print("=="*20)

print("Всі значення Series:")
print(my_series.values)
```

# Завдання 2: Створення DataFrame вручну
- Створи DataFrame, що містить:
    - Імʼя:  Оля    Іван     Марія
    - Вік:   22     35       28
    - Стать: Жінка  Чоловік  Жінка
- Виведи перші 2 рядки
- Виведи останній рядок
- Сформуй випадковий рядок
- Застосуй info() та describe()

```python
import pandas as pd
import numpy as np
#1
data_df = {
    'Ім\'я': ['Оля', 'Іван', 'Марія'],
    'Вік': [22, 35, 28],
    'Стать': ['Жінка', 'Чоловік', 'Жінка']
}
my_dataframe = pd.DataFrame(data_df)
#2
print("Створення DataFrame:")
print(my_dataframe)
print("=="*20)
#3
print("Перші 2 рядки DataFrame:")
print(my_dataframe.head(2))
print("=="*20)
#4
print("Останній рядок DataFrame:")
print(my_dataframe.tail(1))
print("=="*20)
#5
random_index = np.random.choice(my_dataframe.index)
print(f"Випадковий рядок (за індексом {random_index}):")
print(my_dataframe.loc[random_index])
print("=="*20)
#6
print("Інформація про DataFrame (info()):")
my_dataframe.info()
print("=="*20)
#7
print("Описання статистики DataFrame (describe()):")
print(my_dataframe.describe())
```
