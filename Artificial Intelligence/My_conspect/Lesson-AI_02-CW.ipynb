# Завдання 2. Клас “Студент” з підрахунком середнього балу
- Опис: Створи клас Student з атрибутами: name, grades (список оцінок).
    - Додай метод average_grade(), який обчислює середній бал.
    - Додай метод is_passing(), який повертає True, якщо середній бал ≥ 60.
    - Створи список студентів і виведи, хто здає екзамен.

```python
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

    def is_passing(self):
        return self.average_grade() >= 60

student1 = Student("Іван Петренко", [85, 90, 78, 92, 88])
student2 = Student("Марія Сидоренко", [55, 62, 48, 58, 60])
student3 = Student("Олег Ковальчук", [95, 98, 100, 97, 99])
student4 = Student("Анна Бондаренко", [40, 50, 60, 55, 45])

students_list = [student1, student2, student3, student4]

print("Результати складання іспиту:")
for student in students_list:
    avg = student.average_grade()
    if student.is_passing():
        print(f"{student.name}: складає іспит. Середній бал: {avg:.2f}")
    else:
        print(f"{student.name}: НЕ складає іспит. Середній бал: {avg:.2f}")
```

# Завдання 3. Клас “Банк” з операціями
- Опис: Створи клас BankAccount з атрибутами: owner, balance.
    - Методи:
      - deposit(amount) — додати гроші,
      - withdraw(amount) — зняти гроші, якщо вистачає,
      - display_balance() — показати баланс.
      - Створи кілька рахунків та продемонструй операції.

```python
class BankAccount:
    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self.balance = balance
        print(f"Створено новий рахунок для {self.owner} з балансом {self.balance} грн.")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Рахунок поповнено на {amount} грн. Новий баланс: {self.balance} грн.")
        else:
            print("Помилка: сума поповнення повинна бути більше нуля.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Помилка: сума для зняття повинна бути більше нуля.")
        elif amount > self.balance:
            print(f"Помилка: недостатньо коштів. На вашому рахунку лише {self.balance} грн.")
        else:
            self.balance -= amount
            print(f"Знято {amount} грн. Залишок на рахунку: {self.balance} грн.")

    def display_balance(self):
        print(f"Поточний баланс для {self.owner}: {self.balance} грн.")


# --- Приклади ---

print("--- Рахунок Петра ---")
clt1 = BankAccount("Петро Іваненко", 1000.0)
clt1.display_balance()
clt1.deposit(500.50)
clt1.withdraw(200.0)
clt1.withdraw(2000.0)    #Щоб провірити помилку
clt1.deposit(-100)    #Щоб провірити помилку
clt1.display_balance()

print("\n" + "="*25 + "\n")

print("--- Рахунок Ольги ---")
clt2 = BankAccount("Ольга Василенко")
clt2.display_balance()
clt2.withdraw(100.0)    #Щоб провірити помилку
clt2.deposit(3000.0)
clt2.display_balance()
```

# Завдання 5. Контейнер для об’єктів: список об’єктів
- Опис:
    - Створи клас Library, який зберігає список об’єктів Book (з Завдання 1).
    - Методи:
        - add_book(book) — додати книгу,
        - list_books() — вивести всі книги,
        - find_by_author(author) — знайти книги певного автора.
    - Продемонструй роботу з кількома книгами.

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def display_info(self):
        print(f"  - Назва: '{self.title}', Автор: {self.author}, Рік: {self.year}")

class Library:
    def __init__(self, name):
        self.name = name
        self.books = []
        print(f"Бібліотеку '{self.name}' створено!")

    def add_book(self, book):
        if isinstance(book, Book):
            self.books.append(book)
            print(f"Книгу '{book.title}' додано до бібліотеки.")
        else:
            print("Помилка: можна додавати лише об'єкти класу Book.")
            
    def list_books(self):
        print(f"\n--- Книги в бібліотеці '{self.name}' ---")
        if not self.books:
            print("Бібліотека порожня.")
        else:
            for book in self.books:
                book.display_info()
        print("----------------------------------------")

    def find_by_author(self, author_name):
        print(f"\n--- Результати пошуку за автором '{author_name}' ---")
        found_books = []
        for book in self.books:
            if book.author == author_name:
                found_books.append(book)       
        if not found_books:
            print("Книг цього автора не знайдено.")
        else:
            for book in found_books:
                book.display_info()
        print("----------------------------------------")


# --- Демонстрація ---
my_library = Library("Моя бібліотека")

book1 = Book("Дюна", "Айзек Азімов", 1965)
book2 = Book("Фундація", "Іван Франко", 1951)
book3 = Book("Темна вежа: Стрілець", "Стівен Кінг", 1982)
book4 = Book("Стар Трек", "Джин Родденберрі", 1966)

my_library.add_book(book1)
my_library.add_book(book2)
my_library.add_book(book3)
my_library.add_book(book4)

my_library.list_books()

my_library.find_by_author("Айзек Азімов")
my_library.find_by_author("Ілон Маск")    #Щоб провірити помилку
```

# Завдання 6. Клас “Черга”
- Опис: Створи клас Queue з методами:
    - enqueue(item) — додати елемент,
    - dequeue() — видалити перший елемент,
    - peek() — подивитися перший елемент.
- Використовуй список як внутрішню структуру.
- Продемонструй чергу із 5 елементів.

```python
class Queue:
    def __init__(self):
        self._items = []

    def is_empty(self):
        return len(self._items) == 0

    def enqueue(self, item):
        self._items.append(item)
        print(f"У черзі: '{item}'.   \tЧерга: {self._items}")

    def dequeue(self):
        if not self.is_empty():
            item = self._items.pop(0)
            print(f"Лікар оглянув: '{item}'.\tЧерга: {self._items}")
            return item
        else:
            print("Черга порожня!")
            return None

    def peek(self):
        if not self.is_empty():
            return self._items[0]
        else:
            return None
    
    def size(self):
        return len(self._items)


# --- Демонстрація ---
print("--- Створюємо чергу пацієнтів ---")
patient_queue = Queue()

print("\n--- Реєстрація пацієнтів в чергу ---")
patient_queue.enqueue("Дональд Трамп")
patient_queue.enqueue("Ілон Маск")
patient_queue.enqueue("Біл Гейтс")
patient_queue.enqueue("Джозеф Безос")
patient_queue.enqueue("Дж.Д.Венс")

print(f"\nЗараз у черзі {patient_queue.size()} пацієнтів.")
first_patient = patient_queue.peek()
print(f"Перший на прийом: {first_patient}")

print("\n--- Лікар починає прийом  ---")
patient_queue.dequeue()
patient_queue.dequeue()

print(f"\nЗараз у черзі {patient_queue.size()} пацієнтів.")
print(f"Наступний на прийом: {patient_queue.peek()}")

print("\n--- Завершення прийому ---")
patient_queue.dequeue()
patient_queue.dequeue()
patient_queue.dequeue()

print("\n--- Спроба викликати наступного ---")
patient_queue.dequeue()    #Щоб провірити помилку
```

# Завдання 7. Клас “Стек”
- Опис: Створи клас Stack з методами:
    - push(item) — додати елемент,
    - pop() — видалити останній елемент,
    - peek() — подивитися останній елемент.
    - Використовуй список як внутрішню структуру.
    - Продемонструй роботу на прикладі чисел.

```python
class Stack:
    def __init__(self):
        self._items = []

    def is_empty(self):
        return len(self._items) == 0

    def push(self, item):
        self._items.append(item)
        print(f"Додавання: {item}.   \tСтек: {self._items}")

    def pop(self):
        if not self.is_empty():
            item = self._items.pop()
            print(f"Видалення: {item}. \tСтек: {self._items}")
            return item
        else:
            print("Видалення:  Стек порожній.")
            return None

    def peek(self):
        if not self.is_empty():
            return self._items[-1]
        else:
            print("Показати не має чого.")
            return None
    
    def size(self):
        return len(self._items)


# --- Демонстрація ---
number_stack = Stack()

print("\n--- Додаємо числа в стек ---")
number_stack.push(10)
number_stack.push(20)
number_stack.push(30)
number_stack.push(40)

print(f"\nЗараз у стеку {number_stack.size()} елементів.")
top_element = number_stack.peek()
print(f"Верхній елемент : {top_element}")

print("\n--- Видаляємо числа зі стеку ---")
number_stack.pop()
number_stack.pop()

print(f"\nЗараз у стеку {number_stack.size()} елементів.")
print(f"Новий верхній елемент: {number_stack.peek()}")

print("\n--- Очищуємо стек ---")
number_stack.pop()
number_stack.pop()

print("\n--- Спроба видалити з порожнього стеку ---")
number_stack.pop()    #Щоб провірити помилку

print(f"\nЗараз у стеку {number_stack.size()} елементів.")
print(f"Новий верхній елемент: {number_stack.peek()}")
```

# Завдання 8. Клас “Гра: Вгадай число”
- Опис: Створи клас GuessGame з методами:
    - start_game() — генерує випадкове число від 1 до 100,
    - make_guess(number) — приймає число та повертає підказку: “Більше”, “Менше”, “Вгадано!”,
    - play() — дозволяє гравцю вводити числа у циклі, поки не вгадає.

```python
import random

class GuessGame:  
    def __init__(self, min_range=1, max_range=100):
        self._min_range = min_range
        self._max_range = max_range
        self._secret_number = None
        self._attempts = 0

    def start_game(self):
        self._secret_number = random.randint(self._min_range, self._max_range)
        self._attempts = 0
        print("="*30)
        print(f"Гра почалася! Я загадав число від {self._min_range} до {self._max_range}.")
        print("Спробуйте його вгадати.")
        print("="*30)

    def make_guess(self, number):
        self._attempts += 1        
        if number < self._secret_number:
            return "Більше"
        elif number > self._secret_number:
            return "Менше"
        else:
            return "Вгадав!"

    def play(self):
        self.start_game()        
        while True:
            try:
                guess = int(input("Введіть число: "))
            except ValueError:
                print("Помилка: введіть коректне число.")
                continue            
            result = self.make_guess(guess)
            print(f"Підказка: {result}")           
            if result == "Вгадав!":
                break
        
        print("-" * 30)
        print(f" Вітаю! Ви вгадали число {self._secret_number} за {self._attempts} спроб! ")
        print("-" * 30)

# --- Запуск гри ---
if __name__ == "__main__":
    game = GuessGame()
    game.play()
```

# Завдання 9. Клас “Менеджер задач”
- Опис: Створи клас Task з полями: title, description, priority.
    - Створи клас TaskManager, який зберігає список задач (Task):
        - Додати задачу (add_task)
        - Видалити задачу (remove_task)
        - Сортувати задачі за пріоритетом (sort_by_priority)
        - Показати всі задачі (show_tasks)

```python
class Task:
    def __init__(self, title, description, priority):
        self.title = title
        self.description = description
        self.priority = priority

    def __str__(self):
        return f"[Пріоритет: {self.priority}] {self.title}: {self.description}"

class TaskManager:
    def __init__(self):
        self._tasks = []

    def add_task(self, task):
        if isinstance(task, Task):
            self._tasks.append(task)
            print(f" Задача '{task.title}' успішно додана.")
        else:
            print(" Помилка: щось пішло не так.")

    def show_tasks(self):
        print("\n--- Ваш список задач ---")
        if not self._tasks:
            print("Список порожній!")
        else:
            for i, task in enumerate(self._tasks, 1):
                print(f"{i}. {task}")
        print("--------------------------\n")

    def remove_task(self, task_title):
        task_to_remove = None
        for task in self._tasks:
            if task.title == task_title:
                task_to_remove = task
                break       
        if task_to_remove:
            self._tasks.remove(task_to_remove)
            print(f" Задачу '{task_title}' видалено.")
        else:
            print(f" Помилка: Задачу з назвою '{task_title}' не знайдено.")

    def sort_by_priority(self):
        self._tasks.sort(key=lambda task: task.priority, reverse=True)
        print(" Задачі відсортовано за пріоритетом.")


# --- Демонстрація ---
if __name__ == "__main__":
    manager = TaskManager()
    manager.show_tasks()

    task1 = Task("Купити хліба", "Зайти в магазин після роботи", 3)
    task2 = Task("Терміновий дзвінок", "Погасити кредит", 5)
    task3 = Task("Вивчити Jupyter", "Зробити завдання №1-2", 4)
    task4 = Task("Полити квіти", "Не забути про фікус", 1)
    
    manager.add_task(task2)
    manager.add_task(task1)
    manager.add_task(task4)
    manager.add_task(task3)
    manager.show_tasks()

    manager.sort_by_priority()
    manager.show_tasks()

    manager.remove_task("Купити хліба")
    manager.remove_task("Подивитись серіал")    #Щоб провірити помилку
    manager.show_tasks()
```

# Завдання 10. Клас “Менеджер банківських транзакцій”
- Опис: Створи клас TransactionManager, який керує банківськими транзакціями між рахунками.
    - Внутрішні структури:
        - Словник accounts у форматі {account_id: balance}
        - Список transactions для збереження історії транзакцій у форматі {"from": id1, "to": id2, "amount": x}
    - Методи:
        1. add_account(account_id, initial_balance) — створює новий рахунок з початковим балансом
        2. transfer(from_id, to_id, amount) — переводить гроші між рахунками, якщо вистачає коштів; додає запис у список транзакцій
        3. get_balance(account_id) — повертає поточний баланс рахунку
        4. transaction_history(account_id=None) — повертає список всіх транзакцій або тільки для конкретного рахунку
        5. richest_account() — знаходить рахунок із найбільшим балансом

```python
class TransactionManager:
    def __init__(self):
        self._accounts = {}
        self._transactions = []
        print(" Менеджер транзакцій успішно створено.")

    def add_account(self, account_id, initial_balance=0.0):
        if account_id in self._accounts:
            print(f" Помилка: Рахунок з ID '{account_id}' вже існує.")
            return False
        if initial_balance < 0:
            print(" Помилка: Початковий баланс не може бути від'ємним.")
            return False
        self._accounts[account_id] = initial_balance
        print(f" Створено рахунок '{account_id}' з балансом {initial_balance:.2f} грн.")
        return True

    def transfer(self, from_id, to_id, amount):
        if from_id not in self._accounts or to_id not in self._accounts:
            print(f" Транзакція не відбулася: Один з рахунків ('{from_id}' або '{to_id}') не існує.")
            return False
        if amount <= 0:
            print(" Транзакція не вдалася: Сума переказу повинна бути додатною.")
            return False
        if self._accounts[from_id] < amount:
            print(f" Транзакція не вдалася: Недостатньо коштів на рахунку '{from_id}'.")
            return False
        self._accounts[from_id] -= amount
        self._accounts[to_id] += amount
        transaction_record = {"з рахунку": from_id, "на рахунок": to_id, "сума коштів": amount}
        self._transactions.append(transaction_record)
        print(f" Успішний переказ {amount:.2f} грн з '{from_id}' на '{to_id}'.")
        return True

    def get_balance(self, account_id):
        balance = self._accounts.get(account_id)
        if balance is not None:
            return balance
        else:
            print(f" Рахунок '{account_id}' не знайдено.")
            return None

    def transaction_history(self, account_id=None):
        if account_id is None:
            return self._transactions
        if account_id not in self._accounts:
            print(f" Рахунок '{account_id}' не знайдено.")
            return []
        history = [t for t in self._transactions if t['з рахунку'] == account_id or t['на рахунок'] == account_id]
        return history

    def richest_account(self):
        if not self._accounts:
            return None, None
        richest_id = max(self._accounts, key=self._accounts.get)
        max_balance = self._accounts[richest_id]
        return richest_id, max_balance


# --- Демонстрація ---
if __name__ == "__main__":
    manager = TransactionManager()

    print("\n--- 1. Створення рахунків ---")
    manager.add_account("UA001", 10_000)
    manager.add_account("UA002", 500)
    manager.add_account("UA003", 15_000)
    manager.add_account("UA001")    #Щоб провірити помилку

    print("\n--- 2. Перевірка балансів ---")
    print(f"Баланс UA001: {manager.get_balance('UA001'):.2f} грн")
    print(f"Баланс UA003: {manager.get_balance('UA003'):.2f} грн")

    print("\n--- 3. Проведення транзакцій ---")
    manager.transfer("UA001", "UA002", 1_500)
    manager.transfer("UA003", "UA001", 3_000)
    manager.transfer("UA002", "UA003", 5_000)    #Щоб провірити помилку

    print("\n--- 4. Перевірка балансів після транзакцій ---")
    print(f"Новий баланс UA001: {manager.get_balance('UA001'):.2f} грн")
    print(f"Новий баланс UA002: {manager.get_balance('UA002'):.2f} грн")
    print(f"Новий баланс UA003: {manager.get_balance('UA003'):.2f} грн")
    
    print("\n--- 5. Історія транзакцій ---")
    print("Всі транзакції:")
    all_history = manager.transaction_history()
    for trans in all_history:
        print(f"  - {trans}")
        
    print("\nІсторія для рахунку UA001:")
    ua001_history = manager.transaction_history("UA001")
    for trans in ua001_history:
        print(f"  - {trans}")

    print("\n--- 6. Пошук найбагатшого клієнта ---")
    acc_id, balance = manager.richest_account()
    if acc_id:
        print(f" Найбільший баланс на рахунку '{acc_id}': {balance:.2f} грн.")
```
