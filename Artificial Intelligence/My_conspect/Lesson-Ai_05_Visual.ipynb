## **Побудова графіків функцій**

```python
import numpy as np              
import matplotlib.pyplot as plt 

x = np.linspace(-10, 10, 100)

y = x ** 2

plt.figure(figsize=(8, 5))

plt.plot(x, y, label='y = x^2', color='blue')

plt.title('Графік функції y = x^2')

plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```

## **Побудова гістограм розподілу змінних**

```python
import numpy as np             
import matplotlib.pyplot as plt 

data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(8, 5))

plt.hist(data, bins=30, color='skyblue', edgecolor='black')

plt.title('Гістограма нормального розподілу')

plt.xlabel('Значення')
plt.ylabel('Кількість')

plt.show()
```

## **Кругові діаграми (Pie charts)**

```python
import matplotlib.pyplot as plt 

labels = ['Яблука', 'Банани', 'Вишні', 'Апельсини']
sizes = [30, 20, 25, 25]  

explode = (0, 0.1, 0, 0)

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', startangle=90, shadow=True)

plt.title('Розподіл фруктів у кошику')

plt.axis('equal')

plt.show()
```

## **Побудова bar-plot діаграм (стовпчикових діаграм)**

```python
import matplotlib.pyplot as plt  

categories = ['Яблука', 'Банани', 'Вишні', 'Апельсини']

values = [25, 30, 15, 20]

x_pos = range(len(categories))

plt.figure(figsize=(8, 5))

plt.bar(x_pos, values, color='orange', edgecolor='black')

plt.xlabel('Фрукти')
plt.ylabel('Кількість')

plt.xticks(x_pos, categories)

plt.title('Кількість фруктів за видами')

plt.show()
```
