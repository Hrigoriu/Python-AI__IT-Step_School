## *MAE — Mean Absolute Error (Середня абсолютна помилка)*

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_true = np.array([15, 17, 19, 21, 24])

y_pred = np.array([14, 18, 17, 20, 22])

mae = mean_absolute_error(y_true, y_pred)
print("MAE:", mae)
```

## *MSE — Mean Squared Error (Середньоквадратична помилка)*

```python
mse = mean_squared_error(y_true, y_pred)
print("MSE:", mse)
```

## *R² — Coefficient of Determination (Коефіцієнт детермінації)*

```python
r2 = r2_score(y_true, y_pred)
print("R²:", r2)
```

## *Функція втрат (Loss function) Функція помилок (Cost function / Error function)*

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_true = np.array([15, 17, 19])
y_pred = np.array([14, 18, 17])

loss_first = (y_true[0] - y_pred[0])**2
print("Loss (1-й приклад, MSE):", loss_first)

cost = np.mean((y_true - y_pred)**2)
print("Cost (MSE):", cost)

mae = mean_absolute_error(y_true, y_pred)
mse = mean_squared_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print("MAE (Mean Absolute Error):", mae)
print("MSE (Mean Squared Error):", mse)
print("R² (коефіцієнт детермінації):", r2)
```

## *Лінійна регресія*

```python
import numpy as np
import matplotlib.pyplot as plt

experience = np.array([1, 2, 3, 4, 5])
salary = np.array([30, 35, 45, 50, 55])

x_mean = np.mean(experience)  
y_mean = np.mean(salary)     

cov_matrix = np.cov(experience, salary, ddof=0)
cov_xy = cov_matrix[0][1]  

var_x = np.var(experience, ddof=0)  

w = cov_xy / var_x      
b = y_mean - w * x_mean 

print(f"Рівняння прямої: salary = {w:.2f} * experience + {b:.2f}")
```

```python
def predict(x):
    return w * x + b

new_exp = np.array([6, 7])
predicted_salary = predict(new_exp)

plt.scatter(experience, salary, label='Фактичні дані')
plt.plot(experience, [predict(x) for x in experience], color='red', label='Лінія регресії')
plt.scatter(new_exp, predicted_salary, color='green', label='Прогноз')
plt.xlabel('Роки досвіду')
plt.ylabel('Зарплата (тис. грн)')
plt.title('Прогноз зарплати за роками досвіду')
plt.legend()
plt.show()
```

## *Приклад лінійної регресії*


```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X = np.array([1, 2, 3])  
y = np.array([15, 17, 19])

model = LinearRegression()  
model.fit(X, y)             

print("Коефіцієнт (w):", model.coef_)
print("Зсув (b):", model.intercept_)

y_pred = model.predict(X)
print("Прогнози:", y_pred)

mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("R²:", r2)
```

## *Розглянемо приклад регресії з кількома змінними (мультифакторна лінійна регресія)*

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([
    [2, 6, 80],
    [3, 7, 90],
    [4, 6, 85],
    [5, 8, 95],
    [3, 5, 70]
])

y = np.array([65, 70, 75, 90, 60])

model = LinearRegression()
model.fit(X, y)

print(f"Коефіцієнти: study_hours = {model.coef_[0]:.2f}, sleep_hours = {model.coef_[1]:.2f}, attendance_rate = {model.coef_[2]:.2f}")
print(f"Інтерсепт (b): {model.intercept_:.2f}")
    
new_student = np.array([[4, 7, 85]])
predicted_grade = model.predict(new_student)
print(f"Прогнозована оцінка: {predicted_grade[0]:.2f}")
```
